## Matplotlib workflow example

```python
#1 import matplotlib

%matplotlib inline
import matplotlib.pyplot as plt

#2 Prepare data

x = [1.5,2.75,3.5,4.8]
y = [15,25,35,45]

#3 Setup plot

fig, ax = plt.subplots(figsize = (5,5)) #(width,height)

#4 Plot Data

ax.plot(x,y)

#5 Customize plot

ax.set(title = "Test plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

#6 Save & show

fig.savefig("test-plot.png")
```

```python
# Trying with another data set

hd= pd.read_csv("041 heart-disease.csv")
hd.head(7)
```

```python
hd["age"].plot.hist();
```

```python
hd.plot.hist(figsize=(10,25),subplots=True);
```

```python
hd.head()
```

```python
over_50 = hd[hd["age"]>50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target");
```

```python
# OO method

fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
             x = "age",
             y = "chol",
             c = "target",
             ax=ax);
```

```python
## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

ax.set(title="Heart Disease & Cholesterol levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Adding a legend
ax.legend(*scatter.legend_elements() , title="Target");

# Adding a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# Subplot of chol, age and thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True);

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

ax0.set(title="Heart Disease & Cholesterol levels",
       ylabel="Cholesterol");

ax0.legend(*scatter.legend_elements() , title="Target");

ax0.axhline(over_50["chol"].mean(),
           linestyle="--");

#Add data to ax1

scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"]);

ax1.set(title="Heart Disease & Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");

ax1.legend(*scatter.legend_elements() , title="Target");

ax1.axhline(over_50["thalach"].mean(),
           linestyle="--");
#Adding a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=15, fontweight="bold");
```

```python
#Set style
plt.style.use("seaborn-whitegrid")

## OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="RdBu");

ax.set(title="Heart Disease & Cholesterol levels",
       xlabel="Age",
       ylabel="Cholesterol");

# Adding a legend
ax.legend(*scatter.legend_elements() , title="Target");

# Adding a horizontal line
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# Subplot of chol, age and thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True);

#Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter");

ax0.set(title="Heart Disease & Cholesterol levels",
       ylabel="Cholesterol");
ax0.set_xlim([50,80])

ax0.legend(*scatter.legend_elements() , title="Target");

ax0.axhline(over_50["chol"].mean(),
           linestyle="--");

#Add data to ax1

scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["thalach"],
                     c=over_50["target"],
                     cmap="winter");

ax1.set(title="Heart Disease & Max Heart Rate",
       xlabel="Age",
       ylabel="Max Heart Rate");
ax1.set_ylim([60,200])

ax1.legend(*scatter.legend_elements() , title="Target");

ax1.axhline(over_50["thalach"].mean(),
           linestyle="--");
#Adding a title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=25, fontweight="bold");
```
